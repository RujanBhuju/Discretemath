Importing the required library to plot the graph from my house to the highschool.
here my friend's houses are the nodes and the path between the houses are the edges.

```python
import graphviz

Path = graphviz.Digraph('Path')
Path.node('M', 'My House')
Path.node('R', 'Rohit House')
Path.node('J', 'Rojesh House')
Path.node('P', 'Prashant House')
Path.node('K', 'Kumesh House')
Path.node('H', 'Highschool')
```

```python
Path.edges(['MR','MJ','RP','RK','RH','JK','JH','PH','KH','MH'])
Path
```
